# Reading the stocks

```python
ticker=["SPY","MSFT","AAPL","TSLA"]
stocks = yf.download(ticker, start = "2010-01-01",end ="2023-01-01")
stocks.head()
display(stocks.tail())
display(stocks.info())
stocks.to_csv("sushmit_stocksYT.csv")
stocks = pd.read_csv("sushmit_stocksYT.csv")
display(stocks.head())
```

## Playing with index

```python
stocks = pd.read_csv("sushmit_stocksYT.csv",header=[0,1],index_col=[0],parse_dates =[0])
display(stocks.info())
display(stocks.head())
#convert mult index to one tuple
stocks.columns=stocks.columns.to_flat_index()
display(stocks.columns)
display(stocks.head())
stocks.columns=pd.MultiIndex.from_tuples(stocks.columns)
display(stocks.head())
display(stocks.describe())
```

## Plotting the close price

```python
close = stocks.loc[:,'Close'].copy()
close_long_form = close.reset_index().melt('Date',var_name = 'company',value_name = 'price').copy()

alt.Chart(close_long_form).mark_line().encode(x= 'Date',y = 'price', color = 'company').properties(
    width=1000,
    height=600,
    title = 'Closing Price'
)
```
